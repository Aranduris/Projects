```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv('C:/Intro/Project/air-quality-india.csv')
data = data.dropna(axis=1)
print(data)
```

```python
def data_year(data):
    plot1 = data.groupby('Year', as_index=False)['PM2.5'].mean()
    print(plot1)
    sb.barplot(x= 'Year', y ='PM2.5', data=plot1, ci=False)
    plt.title("Year_wise_air_pollution")
    plt.savefig("Year_wise_air_pollution")
    plt.show()

def data_month(data):
    plot1 = data.groupby('Month', as_index=False)['PM2.5'].mean()
    print(plot1)
    sb.barplot(x= 'Month', y ='PM2.5', data=plot1, ci=False)
    plt.title("Month_wise_air_pollution")
    plt.savefig("Month_wise_air_pollution")
    plt.show()

def data_hourly(data):
    plot1 = data.groupby('Hour', as_index=False)['PM2.5'].mean()
    print(plot1)
    sb.barplot(x= 'Hour', y ='PM2.5', data=plot1, ci=False)
    plt.title("hourly_wise_air_pollution")
    plt.savefig("hourly_wise_air_pollution")
    plt.show()

data_year(data)
data_month(data)
data_hourly(data)



    
```

```python
sb.heatmap(data.groupby(['Month','Year'])['PM2.5'].mean().unstack(), annot=True, fmt=".0f", cmap = sb.cm.rocket_r, cbar_kws={'label': 'pollution level'})
plt.title("Cumulative air pollution level of India 2017-2022")
plt.savefig("Heatmap")
```
